# Ajout de coordonnées x et y en fonction des postes occupés

```python
positions_mapping = {
    'gardien de but': {'X': 10, 'Y': 50},
    'défenseur': {'X': 30, 'Y': 50},
    'défenseur, attaquant': {'X': 35, 'Y': 50},
    'défenseur, milieu de terrain': {'X': 40, 'Y': 50},
    'milieu de terrain, défenseur': {'X': 45, 'Y': 50},
    'milieu de terrain': {'X': 50, 'Y': 50},
    'milieu de terrain, attaquant': {'X': 55, 'Y': 50},
    'attaquant, milieu de terrain': {'X': 60, 'Y': 50},
    'attaquant, défenseur': {'X': 65, 'Y': 50},
    'attaquant': {'X': 70, 'Y': 50},
}

def assign_coordinates(row):
    poste = row['Poste'].lower().strip()
    if poste in positions_mapping:
        return positions_mapping[poste]['X'], positions_mapping[poste]['Y']
    return None, None

df['X'], df['Y'] = zip(*df.apply(assign_coordinates, axis=1))
```
